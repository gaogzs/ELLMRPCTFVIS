```python
import pandas as pd

hanna_file_dir = "hanna_stories_annotations.csv"
df_hanna = pd.read_csv(hanna_file_dir)
df_stories = df_hanna[["Story ID", "Story"]].drop_duplicates(subset=["Story ID"]).rename(
    columns={"Story ID": "narrative_id", "Story": "narrative"}
)
df_annotations = df_hanna[["Story ID", "Coherence"]].rename(
    columns={"Story ID": "narrative_id", "Coherence": "consistency"}
)

df_stories.to_csv("hanna_stories.csv", index=False)
df_annotations.to_csv("hanna_annotations.csv", index=False)
```

```python
import pandas as pd

hanna_file_dir = "hanna_stories_annotations.csv"
df_hanna = pd.read_csv(hanna_file_dir)
df_annotations = df_hanna[["Story ID", "Coherence"]].rename(
    columns={"Story ID": "narrative_id", "Coherence": "consistency"}
)
df_annotations = df_annotations.groupby("narrative_id").mean().reset_index()
df_annotations.to_csv("hanna_annotations_average.csv", index=False)
```

```python
import pandas as pd

hanna_file_dir = "hanna_stories_annotations.csv"
df_hanna = pd.read_csv(hanna_file_dir)
df_annotations = df_hanna[["Story ID", "Coherence"]].rename(
    columns={"Story ID": "narrative_id", "Coherence": "consistency"}
)
df_annotations = df_annotations.groupby("narrative_id").median().reset_index()
df_annotations.to_csv("hanna_annotations_median.csv", index=False)
```

```python
import pandas as pd

hanna_file_dir = "hanna_stories_annotations.csv"
df_hanna = pd.read_csv(hanna_file_dir)
df_hanna_pivot = df_hanna.pivot_table(
    index="Story ID",
    columns="Worker ID",
    values="Coherence",
    aggfunc="mean",
).reset_index()
col_table =  {"Story ID": "narrative_id", "Coherence": "consistency"}
i = 0
for value in df_hanna["Worker ID"].value_counts().index:
    i += 1
    col_table[value] = f"human_score_{i}"
df_hanna_pivot.rename(columns=col_table, inplace=True)
df_hanna_pivot = df_hanna_pivot[["narrative_id"] + [f"human_score_{i}" for i in range(1, i + 1)]]
df_hanna_pivot.to_csv("hanna_annotations_pivot.csv", index=False)
```

```python
import pandas as pd

swb_file_dir = "story_writing_benchmark_average.csv"
df_swb = pd.read_csv(swb_file_dir)
df_extracted = df_swb[df_swb["language"] == "en"][["story_text"]].drop_duplicates(subset=["story_text"]).rename(
    columns={"story_text": "narrative"}
)
df_extracted.reset_index(inplace=True)
df_extracted.index.name = "narrative_id"
df_extracted = df_extracted[["narrative"]]
df_extracted.to_csv("swb_stories.csv", index=True)
df_extracted
```

```python
import pandas as pd

swb_file_dir = "story_writing_benchmark_average.csv"
df_swb = pd.read_csv(swb_file_dir)
df_good_stories = df_swb[(df_swb["language"] == "en") & (df_swb["q2"] >= 3.0) & (df_swb["q5"] >= 3.0) & (df_swb["q6"] >= 3.0) & (df_swb["q7"] >= 3.0)].reset_index()
df_good_stories.index.name = "narrative_id"
df_good_stories = df_good_stories[["story_text"]].rename(columns={"story_text": "narrative"})
df_good_stories.to_csv("swb_good_stories.csv", index=True)
df_good_stories
```
